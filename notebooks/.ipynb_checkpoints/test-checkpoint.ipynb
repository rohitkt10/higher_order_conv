```python
import numpy as np

import tensorflow as tf
from tensorflow import keras as tfk

from pairwise_conv_1d import PairwiseConv1D

import matplotlib.pyplot as plt 
import matplotlib as mpl
import seaborn as sns 
sns.set()
%matplotlib inline
```

```python
kwargs = {'filters':32, 'kernel_size':19, 
          'kernel_regularizer':tfk.regularizers.l2(1e-6), 
          'padding':'same'}
conv1 = tfk.layers.Conv1D()

conv2 = PairwiseConv1D(filters=32, kernel_size=19, 
                       kernel_regularizer=tfk.regularizers.l2(1e-6),
                      padding='same', name='conv2', use_bias=True)

L, A = 50, 4
x = tf.random.normal((2, L, A))
y1 = conv1(x)
y2 = conv2(x)
print(y1.shape, y2.shape)
```

```python
print(conv2.diag_kernel.shape, conv1.kernel.shape)
```

```python
diag = tf.transpose(conv2.diag_kernel, [1, 0])
diag = tf.linalg.diag(diag) 
print(diag[0].numpy())
```

```python
w = conv2.kernel.numpy()[:,:,0]
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(w, linewidths=1, cbar=True)
```
